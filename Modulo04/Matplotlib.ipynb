```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.plot(x,y, marker ='o',color='#99d8c9',linestyle='--',label='linea 1')  #colores extraidos de https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3
plt.plot(y,x, marker ='o',color='#2ca25f',linestyle='--',label='linea 2')

plt.title('Este es el titulo')

plt.xlabel('Edad')
plt.ylabel('Nota')

plt.legend()
plt.grid(True)

plt.show()
```
