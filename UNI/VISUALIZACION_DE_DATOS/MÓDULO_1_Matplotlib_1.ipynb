## **<font color="#000000">Importancia de la visualización de datos</font>**

</td>
</tr>
</table>

>* **Facilita la Comprensión de los Datos**
La visualización transforma datos complejos en representaciones gráficas, lo que permite a los analistas y a los interesados comprender mejor la información. Los gráficos pueden revelar patrones, tendencias y anomalías que no son evidentes en tablas de datos.
>* **Comunicación Clara de Resultados**
Las visualizaciones son herramientas efectivas para comunicar hallazgos a audiencias no técnicas. Un gráfico bien diseñado puede explicar más que un extenso informe escrito, facilitando la toma de decisiones informadas.
>* **Identificación de Tendencias y Patrones**
La visualización ayuda a identificar tendencias y patrones en los datos a lo largo del tiempo. Por ejemplo, gráficos de líneas pueden mostrar cómo varían las ventas a lo largo de varios meses, permitiendo a las empresas anticipar cambios en el mercado.
>* **Análisis Exploratorio de Datos (EDA)**
Antes de aplicar modelos estadísticos o de machine learning, es crucial realizar un análisis exploratorio de datos. Las visualizaciones permiten a los analistas explorar la distribución de los datos, identificar outliers y entender la relación entre diferentes variables.
>* **Soporte para Toma de Decisiones**
Al presentar datos visualmente, las organizaciones pueden tomar decisiones más rápidas y precisas. Los dashboards interactivos permiten a los ejecutivos monitorear KPIs y otros indicadores clave de rendimiento de manera eficiente.
>* **Flexibilidad y Personalización**
Python ofrece una variedad de bibliotecas de visualización, como **Matplotlib**, **Seaborn** y **Plotly**, que permiten la creación de gráficos personalizados. Esto significa que los usuarios pueden adaptar las visualizaciones a sus necesidades específicas y preferencias estéticas.
>* **Integración con Análisis de Datos**
Python es una herramienta popular en el análisis de datos debido a su capacidad para manejar grandes conjuntos de datos y su compatibilidad con otras bibliotecas de análisis, como Pandas y NumPy. La visualización de datos se integra perfectamente en el flujo de trabajo de análisis de datos.
>* **Interactividad**
Con bibliotecas como **Plotly** y **Bokeh**, es posible crear visualizaciones interactivas que permiten a los usuarios explorar los datos en profundidad. Esto es especialmente útil en entornos de negocios donde las preguntas pueden variar y la exploración de datos es clave.

## **<font color="#000000">Creación de la tabla calendario</font>**

</td>
</tr>
</table>

````python
import numpy as np
import pandas as pd

fecha_min = df_FactResellerSales['OrderDate'].min()
fecha_max = df_FactResellerSales['OrderDate'].max()

fechas = pd.date_range(fecha_min, fecha_max, freq='D')

df_DimCalendario = pd.DataFrame({'Fecha': fechas})

df_DimCalendario['Año'] = df_DimCalendario['Fecha'].dt.year
df_DimCalendario['Semestre'] = np.where(df_DimCalendario['Fecha'].dt.month <= 6, 'S-1', 'S-2')
df_DimCalendario['Trimestre'] = ((df_DimCalendario['Fecha'].dt.month -1) // 3 + 1).apply(lambda x: f'T-{x}')
df_DimCalendario['Mes'] = df_DimCalendario['Fecha'].dt.month_name()
df_DimCalendario['MesAbr'] = df_DimCalendario['Fecha'].dt.month_name().str[:3]
df_DimCalendario['MesNum'] = df_DimCalendario['Fecha'].dt.month
df_DimCalendario['SemAño'] = df_DimCalendario['Fecha'].dt.isocalendar().week
df_DimCalendario['Dia'] = df_DimCalendario['Fecha'].dt.day_name()
df_DimCalendario['DiaAbr'] = df_DimCalendario['Fecha'].dt.day_name().str[:3]
df_DimCalendario['DiaNumSem'] = df_DimCalendario['Fecha'].dt.dayofweek + 1
df_DimCalendario['DiaMes'] = df_DimCalendario['Fecha'].dt.day
````

````python
df_categoria_subcategoria = pd.merge(df_DimProductCategory, df_DimProductSubcategory, on='ProductCategoryKey', how='inner').copy()
df_categoria_subcategoria_producto = pd.merge(df_categoria_subcategoria, df_DimProduct, on='ProductSubcategoryKey', how='inner').copy()
df_categoria_subcategoria_producto = df_categoria_subcategoria_producto.rename(columns={'StartDate': 'FechaIniPrd', 'EndDate': 'FechaFinPrd'}).copy()

df_DimPromotion['MaxQty'] = df_DimPromotion['MaxQty'].fillna(99999)

df_FactVentas = pd.merge(df_FactResellerSales, df_DimPromotion, on='PromotionKey', how='inner').copy()
df_FactVentas = pd.merge(df_FactVentas, df_categoria_subcategoria_producto, on='ProductKey', how='inner').copy()
df_FactVentas = pd.merge(df_FactVentas, df_DimCalendario, left_on='OrderDate', right_on='Fecha', how='inner').copy()

````

### **<font color="#FFFFFF">Costos</font>**

</td>
</tr>
</table>

````python
df_FactVentas['Costos'] = df_FactVentas['OrderQuantity'] * df_FactVentas['ProductStandardCost']
````

<table width='99%'>
<tr>
<td bgcolor='#306998'>

### **<font color="#FFFFFF">Ventas</font>**

</td>
</tr>
</table>

````python
df_FactVentas['Ventas'] = df_FactVentas['OrderQuantity'] * df_FactVentas['UnitPrice'] * (
    1
    -
    np.where
    (
        (df_FactVentas['OrderDate'] >= df_FactVentas['StartDate']) &
        (df_FactVentas['OrderDate'] <= df_FactVentas['EndDate']) &
        (df_FactVentas['OrderQuantity'] >= df_FactVentas['MinQty']) &
        (df_FactVentas['OrderQuantity'] <= df_FactVentas['MaxQty']),
        df_FactVentas['DiscountPct'],
        df_FactVentas['UnitPriceDiscountPct']
    )
    )
````

<table width='99%'>
<tr>
<td bgcolor='#306998'>

### **<font color="#FFFFFF">Utilidad</font>**

</td>
</tr>
</table>

````python
df_FactVentas['Utilidad'] = df_FactVentas['Ventas'] - df_FactVentas['Costos']
````

<table width='99%'>
<tr>
<td bgcolor='#306998'>

### **<font color="#FFFFFF">Porcentaje del margen de utilidad</font>**

</td>
</tr>
</table>

$$\text{Porcentaje del margen de utilidad} = \frac{Utilidad}{Ventas}*100$$

````python
df_FactVentas['PorcentajeMargen'] = (df_FactVentas['Utilidad'] / df_FactVentas['Ventas']) * 100
````

````python
costos= df_FactVentas['Costos'].sum()
ventas = df_FactVentas['Ventas'].sum()
utilidad = ventas - costos

porcentaje_margen = (utilidad / ventas) * 100

print(f'Costos: ${costos:,.2f}')
print(f'Ventas: ${ventas:,.2f}')
print(f'Utilidad Neta: ${utilidad:,.2f}')
print(f'Porcentaje del margen de utilidad: {porcentaje_margen:,.2f}%')
````

>* **groupby** es una función extremadamente poderosa y fundamental en Pandas para el análisis de datos. Te permite agrupar filas de un DataFrame en función de los valores de una o más columnas, y luego realizar cálculos sobre cada grupo de forma independiente.

````python
df_ventas_x_año = df_FactVentas.groupby('Año')['Ventas'].sum().reset_index()
df_ventas_x_año.head(50)
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
plt.plot(df_ventas_x_año['Año'], df_ventas_x_año['Ventas'])
plt.xlabel('Año')
plt.ylabel('Ventas')
plt.xticks(df_ventas_x_año['Año'])
plt.title('Ventas por Año')
plt.show()
````
