## **<font color="#000000">¿Qué es Plotly?</font>**

</td>
</tr>
</table>

Plotly es una biblioteca de visualización de datos en Python que permite crear gráficos interactivos y visualizaciones sofisticadas.

Es especialmente útil para el análisis de datos y la presentación visual de información de manera atractiva.

## **<font color="#000000">Características</font>**

</td>
</tr>
</table>

>* **Interactividad**: Los gráficos generados son interactivos, lo que permite a los usuarios explorar los datos mediante zoom, desplazamiento y selección de elementos.
>* **Variedad de Gráficos**: Plotly admite una amplia gama de tipos de gráficos, incluyendo gráficos de líneas, dispersión, barras, mapas, gráficos de superficie, entre otros.
>* **Integración**: Se integra bien con otras bibliotecas de Python como Pandas y NumPy, facilitando el uso de datos en diferentes formatos.
>* **Dash**: Plotly también ofrece Dash, un framework para construir aplicaciones web analíticas que utilizan visualizaciones interactivas.
>* **Exportación**: Los gráficos pueden ser exportados a varios formatos, incluidos PNG, JPEG y PDF, así como ser incrustados en aplicaciones web.
